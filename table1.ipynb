```python
from tableone import TableOne
import pandas as pd
```

```python
other_data=pd.read_excel('1.20.20 DeIdentified_HG_CT_JJM_AS_Clinical Variables_Final.xlsx')
# dropping excluded pts
o_data=other_data.drop([4,7, 11, 13, 23, 25, 29, 36, 41, 44, 47, 61, 62, 63, 64, 68, 76, 88, 94, 96, 
                          108, 133, 135, 140, 142]).reset_index(drop=True)
o_data
```

```python
#defining columns to include
columns = ['Age at Dx', 'BMI-Category', 'Diabetes (y/n)', 'Histology', 'Final Grade', 'Fundal/Extrafundal', 'Node Status (+/-)', 
           'Adjuvant treatment (y/n)', 'Chemo (y/n)', 'RT (y/n)', 'Recurrence (y/n)', 'Time to Recurrence (months)',
           'Deceased (y/n)', 'Overall Survival (Months)']
categorical = ['BMI-Category', 'Diabetes (y/n)', 'Histology', 'Final Grade', 'Fundal/Extrafundal', 'Node Status (+/-)', 
           'Adjuvant treatment (y/n)', 'Chemo (y/n)', 'RT (y/n)', 'Recurrence (y/n)', 'Deceased (y/n)']


# variables for stratification
groupby = ['Recurrence (y/n)']
```

```python
mytable = TableOne(o_data, columns=columns, categorical=categorical, pval = True, groupby = groupby, smd=True,
                  htest_name=True)
```
