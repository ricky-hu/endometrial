```r
#reading data, then running the rank correlation
require('Hmisc')
data <- read.csv("preds.csv")
risk <- unlist(data[1])
time <- unlist(data[2])
event <- unlist(data[3])
results <- rcorr.cens(risk, Surv(time,event))
ci <- pmax(results['C Index'], (1-results['C Index']))
std <- results['S.D.']/2

results <- list(ci,std)
results
df = data.frame(ci, std)
df
write.csv(df, 'ci.csv', row.names = FALSE)
```
